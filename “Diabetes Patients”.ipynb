```python
import pandas as pd
from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
%matplotlib inline 

import warnings
warnings.filterwarnings('ignore')

import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\HP\\Downloads\\Project 2 MeriSKILL (1)\\diabetes.csv")
df
```

```python
plt.figure(figsize = (12,5))
sns.barplot(x = df['Age'], y = df['Glucose'], palette='viridis')
plt.title('The distribution of fasting blood glucose levels')
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
plt.figure(figsize = (12,5))
sns.barplot(x = 'Age', y = 'Insulin',data = df, palette='viridis')
plt.title('The distribution of insulin levels')
plt.ylim(0, 300) 
plt.xticks(rotation=45)
plt.tight_layout()
```

```python
scatter_matrix(df, figsize=(20,20))
plt.show()
```

```python
outcome_counts = df['Outcome'].value_counts()
outcome_counts
```

```python
plt.figure(figsize=(4, 4))
plt.pie(outcome_counts, labels=['Non-Diabetic', 'Diabetic'], autopct='%1.1f%%')
plt.title('Diabetes Outcome Distribution')
plt.show()
```

```python
sns.scatterplot(x = df['DiabetesPedigreeFunction'], y = df['Age'], hue = df['Outcome'])
plt.show()
```

```python
correlation_matrix = df.corr()
correlation_matrix
```

```python
outcome_correlations = correlation_matrix['Outcome'].sort_values(ascending=False)
outcome_correlations
```

```python
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
